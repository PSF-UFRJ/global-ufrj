## Solução Analítica do Problema de Dois Corpos:

Vamos considerar o problema de dois corpos, como o apresentado na figura abaixo:



Vamos considerar a solução do problema para o centro de massa dos dois corpos, representados pelo vetor $\vec{R_0}$. Como não existem forças externas ao sistema, podemos assumir que a aceleração do sistema como um todo é nula, e:

$$\vec{R_0} = \frac{1}{m_1 + m_2}\left(m_1\vec{x_1} + m_2\vec{x_2}\right) = 0$$

E

$$\frac{d^2}{dt^2}\vec{R_0} = 0 \Rightarrow m_1\frac{d^2}{dt^2}\vec{x_1} + m_1\frac{d^2}{dt^2}\vec{x_1} = 0$$

Se assumirmos os valores de $x_1$ e $x_2$ em função de $\vec{R_0}$ e de um vetor $\vec{R}$ associando a posição do primeiro corpo com o centro de massa, teremos tanto $\vec{x_1}$ quanto $\vec{x_2}$ bem definidos:

$$ \vec{x_1} = \vec{R_0} + \vec{R}$$
$$ \vec{x_2} = \vec{R_0} - \frac{m_1}{m_2}\vec{R}$$
$$ \| \vec{R} \| = r $$

E a distância entre os corpos é dada por:

$$ d = \| x_1 - x_2 \| = \frac{m_1 + m_2}{m_2} \| \vec{R} \| = \frac{m_1 + m_2}{m_2} r$$

E assumimos a seguinte notação:

$$ \vec{e_r} = \frac{\vec{R}}{\| \vec{R} \|} = \frac{\vec{R}}{r} $$

Dessa forma, temos:

$$ \vec{F_1} = m_1\vec{a_1} = - \frac{\gamma m_1 m_2}{d^2} \vec{e_r} = - \frac{\gamma m_1 m_2}{\left(\frac{m_1+m_2}{m_2}\right)^2 r^2} \vec{e_r} \Rightarrow m_1\vec{a_1} = - \frac{\gamma m_1 m_2}{\left(\frac{m_1+m_2}{m_2}\right)^2 r^2} \vec{e_r} \Rightarrow \vec{a_1} = - \frac{\gamma  m_2^3}{\left(m_1+m_2\right)^2 r^2} \vec{e_r} $$

Calculamos apenas para $\vec{F_1}$, porque uma vez tido esse valor, o $\vec{F_2}$ é direto.

E podemos assumir:

$$ K = \frac{-Gm_2^3}{\left(m_1 + m_2\right)^2} \Rightarrow \vec{a} = \frac{d^2}{dt^2}\vec{R} = -\frac{K}{r^2} \vec{e_r} = -\frac{K}{r^3} \vec{R}$$

Se abrimos parâmetro a parâmetro o vetor $\vec{a}$, temos:

$$ (a_x, a_y) = -\frac{K}{r^3} (R_x, R_y), \vec{a} = (a_x, a_y), \vec{R} = (R_x, R_y)$$

O que implica:

$$ (a_x, a_y) = (R_x'', R_y'') = -\frac{K}{r^3} (R_x, R_y)$$

Mas devemos lembrar que, no problema apresentado, e sem forças externas envolvidas, o valor de $r$ é uma constante, e o problema se torna duas equações diferenciais ordinárias de segunda ordem praticamente idênticas, diferindo, apenas, dos valores iniciais:

- $a_x = R_x'' = -\frac{K}{r^3} R_x$
- $a_y = R_y'' = -\frac{K}{r^3} R_y$
