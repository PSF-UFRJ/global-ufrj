```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# The function to be integrated:

def u(X, k1 = 0.04, k2 = 10**4, k3 = 3e7, h = 0.1):
    u1 = X[0]
    u2 = X[1]
    u3 = X[2]
    # j = np.array([[-k1, k2*u3, k2*u2],[k1, -k2*u3-2*k3*u2, -k2*u2],[0,2*k3*u2,0]])
    # eigenvalues, _ = np.linalg.eig(j)
    # minLambda = eigenvalues.min()
    # maxLambda = eigenvalues.max()
    # if(maxLambda>1.0e-15):
    #    print("Error in eigenvalues", maxLambda)
    # hOpt = 2 / abs(minLambda)
    # if(hOpt<h):
    #    print("Error in h", h, hOpt)
        
    return np.array([-k1*u1 + k2*u2*u3, k1*u1 - k2*u2*u3 - k3*(u2**2), k3*(u2**2)], dtype='double')

def ux(X, h = 0.1, k1 = 0.04, k2 = 10**4, k3 = 3e7):
   return X + h * u(X, k1, k2, k3, h)

def evaluate_numeric(h1 = 0.1, k1 = 0.04, k2 = 1e4, k3 = 3e7, init_value = np.array([1, 0, 0], dtype='double'), end = 1):
    t_n = np.arange(0, end + h1, h1, dtype='double')
    f = [init_value]
    for n in range(1,len(t_n)):
        f.append(ux(f[n-1], h1, k1, k2, k3))
    return np.vstack((t_n, np.transpose(f)))

def evaluate_numeric_and_do(h1, k1, k2, k3, end, func):
    solution = evaluate_numeric(h1 = h1, k1 = k1, k2 = k2, k3 = k3, end = end)
    t_n = solution[0]
    u1 = solution[1]
    u2 = solution[2]
    u3 = solution[3]
    func(t_n, u1, u2, u3)
```

```python
ig, ax = plt.subplots()

plt.rcParams['text.usetex'] = True
plt.rcParams['figure.figsize'] = [20, 10]

def multiplot(tn, u1, u2, u3):
    plt.plot(tn, u1, label=r'$t_n$ x $u_1$')
    plt.plot(tn, u2, label=r'$t_n$ x $u_2$')
    plt.plot(tn, u3, label=r'$t_n$ x $u_3$')
    
evaluate_numeric_and_do(0.0001, 0.04, 10**4, 3e7, 1000, multiplot)

ax.set_axisbelow(True)
ax.grid(which='major', linestyle='-', linewidth='0.5', color='red')

ax.set_xlabel(r'$u_n$')
ax.set_ylabel(r'$v_n$')
plt.title("Numeric solution of Differential Equation")
plt.legend()

ax = plt.gca()
```
