## Resolver equação de Van der Pol

$$ u_{n+1} = u_n + hv_n
$$

$$ v_{n+1} = v_n - h(\epsilon (u_n^2-1)v_n - u_n
$$

Condições iniciais: u(0) = 1 e v(0) = 0

```python
#Parametros
h = 0.1
e = 0.1
Tf = 100.0
N = int(Tf/h) #Numero de etapas até chegar no Tf

# Definições
t = np.array([0.0])
u = np.array([1.0])
v = np.array([0.0])

for i in range(0,N):    
    v = np.append(v, v[-1] - h*(e*(u[-1]**2 -1)*v[-1] + u[-1]))
    
    u = np.append(u, u[-1] + h*v[-2])
   
    t = np.append(t, t[-1]+h)
    
#Plotagem
plt.subplot(1,2, 1)
plt.plot(t,u)
plt.xlim([0,100])
plt.ylim([-3,3])
plt.title('$\epsilon$ = 0.1 e h = 0.1')

plt.subplot(1,2,2) 
plt.plot(u,v)
plt.xlim([-3,3])
plt.ylim([-3,3])
plt.title('v x u com $\epsilon$ = 0.1');
```
