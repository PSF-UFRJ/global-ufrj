```python
import numpy as np
import matplotlib.pyplot as plt
import math
import sys

# np.set_printoptions(threshold=sys.maxsize)
```

# Exemplo 1 - Apresentação 9, exemplo 2

```python
# Number of steps - the only input:
N = 101

# Generated parameters
h = 1/(N-1)
T = np.zeros((N, N))
F = np.zeros((N))
t = np.zeros((N))
res_anal = np.zeros((N))

f = lambda x: math.sin(math.pi * x)
f_anal = lambda x: np.sin(np.pi*x)/(np.pi**2)
```

```python
# Prepare differences array:
T[0][0] = 2
T[0][1] = -1
t[0] = 0
F[0] = 0
res_anal[0] = f_anal(t[0])
apos = 0
for i in range(1,N-1):
    T[i][apos] = -1
    T[i][apos+1] = 2
    T[i][apos+2] = -1
    t[i] = t[i-1] + h
    F[i] = (h**2) * f(t[i-1])
    res_anal[i] = f_anal(t[i])
    apos += 1
T[-1][-2] = 2
T[-1][-1] = -1
t[-1] = t[-2]+h
F[-1] = 0
res_anal[-1] = f_anal(t[-1])
```

```python
res = np.linalg.solve(T, F)
```

```python
fig, ax = plt.subplots()

plt.plot(t, res, label="u(x) - numeric")
plt.plot(t, res_anal, label="u(x) - analytic")

ax.set_axisbelow(True)
ax.minorticks_on()
ax.grid(which='major', linestyle='-', linewidth='0.5', color='red')
ax.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.title("Numeric solution of Differential Equation")
plt.legend()

ax = plt.gca()
```

## Cálculo dos Erros

```python
absolute_error = np.abs(res - res_anal)

fig, ax = plt.subplots()

plt.plot(t, absolute_error, label="u(x) - numeric")

ax.set_axisbelow(True)
ax.minorticks_on()
ax.grid(which='major', linestyle='-', linewidth='0.5', color='red')
ax.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.title("Numeric solution of Differential Equation")
plt.legend()

ax = plt.gca()
```

```python
accepted_truncation_error = 1e-15
relative_error = absolute_error / (res_anal + accepted_truncation_error) 

fig, ax = plt.subplots()

plt.semilogy(t, relative_error, label="u(x) - numeric")

ax.set_axisbelow(True)
ax.minorticks_on()
ax.grid(which='major', linestyle='-', linewidth='0.5', color='red')
ax.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.title("Numeric solution of Differential Equation")
plt.legend()

ax = plt.gca()
```

# Exemplo 2 - Apresentação 9, exemplo 1

```python
# Number of steps - the only input:
N = 101

# Generated parameters
h = 1/(N+1)
T = np.zeros((N+2, N+2))
F = np.zeros((N+2))
t = np.zeros((N+2))
res_anal = np.zeros((N+2))

f = lambda x: -math.sin(math.pi * x)
f_anal = lambda x: f(x)/(np.pi**2) + x
```

```python
# Prepare differences array:
T[0][0] = 2
T[0][1] = -1
t[0] = 0
F[0] = 0
res_anal[0] = f_anal(t[0])
apos = 0
for i in range(1,N+1):
    T[i][apos] = -1
    T[i][apos+1] = 2
    T[i][apos+2] = -1
    t[i] = t[i-1] + h
    F[i] = (h**2) * f(t[i-1])
    res_anal[i] = f_anal(t[i])
    apos += 1
T[-1][-2] = -1
T[-1][-1] = 2
t[-1] = t[-2]+h
F[-1] = 1
res_anal[-1] = f_anal(t[-1])
```

```python
res = np.linalg.solve(T, F)
```

```python
fig, ax = plt.subplots()

plt.plot(t, res, label="u(x) - numeric")
plt.plot(t, res_anal, label="u(x) - analytic")

ax.set_axisbelow(True)
ax.minorticks_on()
ax.grid(which='major', linestyle='-', linewidth='0.5', color='red')
ax.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.title("Numeric solution of Differential Equation")
plt.legend()

ax = plt.gca()
```

# Cálculo dos Erros

```python
absolute_error = np.abs(res - res_anal)

fig, ax = plt.subplots()

plt.plot(t, absolute_error, label="u(x) - numeric")

ax.set_axisbelow(True)
ax.minorticks_on()
ax.grid(which='major', linestyle='-', linewidth='0.5', color='red')
ax.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.title("Numeric solution of Differential Equation")
plt.legend()

ax = plt.gca()
```

```python
accepted_truncation_error = 1e-15
relative_error = absolute_error / (res_anal + accepted_truncation_error) 

fig, ax = plt.subplots()

plt.semilogy(t, relative_error, label="u(x) - numeric")

ax.set_axisbelow(True)
ax.minorticks_on()
ax.grid(which='major', linestyle='-', linewidth='0.5', color='red')
ax.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.title("Numeric solution of Differential Equation")
plt.legend()

ax = plt.gca()
```
