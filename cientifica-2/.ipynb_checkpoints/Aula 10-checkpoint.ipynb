```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# Number of steps - the only input:
N = 101

# Generated parameters
h = 1/(N-1)
T = np.zeros((N+2, N+2))
F = np.zeros((N+2))
t = np.zeros((N+2))

f = lambda x: math.sin(math.pi * x)
```

```python
# Prepare differences array:
T[0][0] = -1
T[0][1] = 1
t[0] = 0
apos = 0
for i in range(1,N+1):
    T[i][apos] = -1
    T[i][apos+1] = 2
    T[i][apos+2] = -1
    F[i] = (h ** 2) * f((i-1)*h)
    t[i] = t[i-1] + h
    apos += 1
T[-1][-2] = -1 + h
T[-1][-1] = -1
F[-1] = -h
t[-1] = t[-2] + h
```

```python
res = np.linalg.solve(T, F)
```

```python
fig, ax = plt.subplots()

plt.plot(t, res, label="u(x)")
ax.set_axisbelow(True)
ax.minorticks_on()
ax.grid(which='major', linestyle='-', linewidth='0.5', color='red')
ax.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.title("Numeric solution of Differential Equation")
plt.legend()

ax = plt.gca()
```
