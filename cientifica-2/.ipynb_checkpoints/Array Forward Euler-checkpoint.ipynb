```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# The function to be integrated:

def u(X, eps = 0.1):
    u_n = X[0]
    v_n = X[1]
    return np.array([v_n, -eps*(u_n*u_n - 1)*v_n-u_n], dtype='double')

def ux(X, h, eps = 0.1):
   return X + h * u(X, eps)

# Another way to write ux:
# def ux(X, h, eps = 0.1):
#    u_n = X[0]
#    v_n = X[1]
#    return np.array([u_n + h * v_n, v_n + h* (-eps*(u_n*u_n -1)*v_n - u_n)])

def evaluate_numeric(h1 = 0.1, eps = 0.1, init_value = np.array([1, 0], dtype='double'), end = 100):
    t_n = np.arange(0, end + h1, h1, dtype='double')
    f = [init_value]
    for n in range(1,len(t_n)):
        f.append(ux(f[n-1], h1, eps))
    return np.vstack((t_n, np.transpose(f)))

def evaluate_numeric_and_do(h1, eps, func):
    solution = evaluate_numeric(h1 = h1, eps = eps)
    t_n = solution[0][500:]
    u_n = solution[1][500:]
    v_n = solution[2][500:]
    func(t_n, u_n, v_n)
```

```python
ig, ax = plt.subplots()

plt.rcParams['text.usetex'] = True
plt.rcParams['figure.figsize'] = [20, 20]

evaluate_numeric_and_do(0.1, 0.01, lambda tn, u_n, v_n: plt.plot(u_n, v_n, label=r'$u_n$ x $v_n$'))
evaluate_numeric_and_do(0.1, 0.1, lambda tn, u_n, v_n: plt.plot(u_n, v_n, label=r'$u_n$ x $v_n$'))
evaluate_numeric_and_do(0.1, 1, lambda tn, u_n, v_n: plt.plot(u_n, v_n, label=r'$u_n$ x $v_n$'))
evaluate_numeric_and_do(0.1, 1.5, lambda tn, u_n, v_n: plt.plot(u_n, v_n, label=r'$u_n$ x $v_n$'))
evaluate_numeric_and_do(0.01, 2, lambda tn, u_n, v_n: plt.plot(u_n, v_n, label=r'$u_n$ x $v_n$'))
evaluate_numeric_and_do(0.01, 2.5, lambda tn, u_n, v_n: plt.plot(u_n, v_n, label=r'$u_n$ x $v_n$'))
evaluate_numeric_and_do(0.01, 3, lambda tn, u_n, v_n: plt.plot(u_n, v_n, label=r'$u_n$ x $v_n$'))
evaluate_numeric_and_do(0.01, 3.5, lambda tn, u_n, v_n: plt.plot(u_n, v_n, label=r'$u_n$ x $v_n$'))
evaluate_numeric_and_do(0.01, 4, lambda tn, u_n, v_n: plt.plot(u_n, v_n, label=r'$u_n$ x $v_n$'))

ax.set_axisbelow(True)
ax.grid(which='major', linestyle='-', linewidth='0.5', color='red')

ax.set_xlabel(r'$u_n$')
ax.set_ylabel(r'$v_n$')
plt.title("Numeric solution of Differential Equation")
plt.legend()

ax = plt.gca()
```
