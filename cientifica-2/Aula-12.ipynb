```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 9
h = 1.0 / (N+1)
T_L,T_R,T_T,T_B = 0.0,0.0,0.0,0.0
A = np.zeros((N*N,N*N))
f = np.zeros((N*N))

for j in range(0,N):
    for i in range(0,N):
        xi, yj = h * i, h * j
        n = i + (j-1) * N - 1
        A[n,n] = 4
        f[n] = h * h * 1.0
        
        if (i>0):
            A[n,n-1] = -1
        else:
            f[n] = f[n] + T_L
        if (i==N):
            f[n] = f[n] + T_R
        else:
            A[n, n+1] = -1
            
        if (j>1):
            A[n, n-N] = -1
        else:
            f[n] = f[n] + T_B
        if (j==N):
            f[n] = f[n] + T_T
        else:
            A[n,n+N] = -1
        
```

```python
A
```

```python
u = np.linalg.solve(A, f)
```

```python
u = u.reshape((N,N))
plt.imshow(u, interpolation='bilinear')
plt.colorbar()
```
