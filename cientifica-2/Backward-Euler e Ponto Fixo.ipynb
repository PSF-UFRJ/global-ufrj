```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

# Declarations

* Fixed point function definition (`fixed_point`)
* Numeric Backward Euler step (`ux`)

```python
def fixed_point(x0, fx, eps, max_loop = 100000):
    stop_point_val = float('inf')
    x = x0
    prev_fx = x0
    for i in range(max_loop):
        if stop_point_val < eps: break
        prev_fx = x
        x = fx(x)
        stop_point_val = abs(x - prev_fx)
    return x, stop_point_val
```

```python
# Numeric ODE Solution:
def ux(t_n, u_n1 = 1, h1 = 0.1, err = 1e-6):
    x, final_error = fixed_point(u_n1, lambda u: u_n1 + h1*(u-t_n)/(u+t_n), err)
    # Discard final error value (unnecessary here)
    return x

## Exact solution for approximation by fixed point (version 1)
def u1(t, u_1):
    return math.sqrt(math.exp(3.1416-2*math.atan(u_1/t))-t**2)

## Exact solution for approximation by fixed point (version 2):
def u2(t, u_1):
    return t * math.tan((3.1416-math.log(t**2+u_1**2))/2)

def u(t_n, u_n1 = 1, h1 = 0.1, err = 1e-6):
    try:
        x, final_error = fixed_point(u_n1, lambda u: u1(t_n, u), err)
        evaluate_error = True
    except:
        x, final_error = fixed_point(u_n1, lambda u: u2(t_n, u), err)
        evaluate_error = False
    if evaluate_error and final_error > err:
        x, final_error = fixed_point(u_n1, lambda: u2(t_n, u), err)
    # Discard final error value (unnecessary here)
    return x
```

## Evaluating Differential Equation Numerically

```python
def evaluate_numeric(h1 = 0.1, end = 7):
    t_n = np.arange(0, end + h1, h1)
    u_n = [1]
    for i in range(1,len(t_n)):
        u_n.append(ux(t_n[i], u_n[i-1], h1))
    return t_n, u_n

# Evaluate exact solution of u(t), using fixed point:
def evaluate_exact(h1 = 0.1, end = 7):
    t_n = np.arange(0, end + h1, h1)
    u_n = [1]
    for i in range(1,len(t_n)):
        u_n.append(u(t_n[i], u_n[i-1]))
    return t_n, u_n 
```

## Plotting - h = 0.1

```python
t_n, u_n = evaluate_numeric()
et_n, eu_n = evaluate_exact()

fig, ax = plt.subplots()

plt.plot(t_n, u_n, label="u(x)")
plt.plot(et_n, eu_n, label="exact u(x)")
ax.set_axisbelow(True)
ax.minorticks_on()
ax.grid(which='major', linestyle='-', linewidth='0.5', color='red')
ax.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.title("Numeric solution of Differential Equation")
plt.legend()

ax = plt.gca()
```
