```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def derivada_lateral(f, x, h):
    return ( f(x+h) - f(x) )/h
```

```python
def derivada_central(f, x, h):
    return ( f(x+h) - f(x-h) )/(2*h)
```

# Gráfico do erro em função de h

## Vários pontos e várias funções diferentes

```python
points = np.arange(0, 2 * np.pi, 0.1)
func = np.exp
dfunc = np.exp

def ndfunc_c(x, h=1e-6):
    return derivada_central(func, x, h)

def ndfunc_l(x, h=1e-6):
    return derivada_lateral(func, x, h)

ydfunc = dfunc(points)
yndfunc_c = ndfunc_c(points)
yndfunc_l = ndfunc_l(points)

plt.plot(points, ydfunc, label="Analytic Derivative")
plt.plot(points, yndfunc_c, label="Numeric Central Derivative")
plt.plot(points, yndfunc_c, label="Numeric Lateral Derivative")
plt.legend()
```

```python
central_error = np.abs(ydfunc - yndfunc_c)

plt.plot(points, central_error, label="Central Derivative Error")
plt.legend()
```

```python
lateral_error = np.abs(ydfunc - yndfunc_l)

plt.plot(points, lateral_error, label="Lateral Derivative Error")
plt.legend()
```

```python
# step = 1e-10
# hh = np.arange(step, 1e-8, step)

hh = np.logspace(-53,0,base=2,num=10000)

# List comprehension is not necessary:
# def errors_c(hs, x=1.6):
#     return np.abs(np.array([dfunc(x) - ndfunc_c(x, h) for h in hs]))
# def errors_l(hs, x=1.6):
#     return np.abs(np.array([dfunc(x) - ndfunc_l(x, h) for h in hs]))

def errors_c(hs, x=1.6):
    return np.abs(dfunc(x) - ndfunc_c(x, hs))
def errors_l(hs, x=1.6):
    return np.abs(dfunc(x) - ndfunc_l(x, hs))
```

```python
newx = 1
yerrors_c = errors_c(hh, newx)
plt.loglog(hh, yerrors_c, label="Erros Derivada Central")
```

```python
yerrors_l = errors_l(hh, newx)
plt.loglog(hh, yerrors_l, label="Erros Derivada Lateral")
```
