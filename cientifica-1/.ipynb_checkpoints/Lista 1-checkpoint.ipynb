## Quest√£o 1 .1 - Resposta:

Podemos considerar uma fun√ß√£o $f$, e uma tangente a ela, conforme a figura abaixo:


Pela figura, vemos que a tangente passa por $f$ no ponto $(x_0, f(x_0) = y_0)$ e que o ponto $x_1$ pode ser encontrado por:

$$
\frac{\Delta y}{\Delta x} = \frac{y_0 - 0}{x_0 - x_1} = f'(x_0) \Rightarrow x_1 = x_0 - \frac{f(x_0)}{f'(x_0)}
$$

Como temos que aplicar reiteiradamente o m√©todo, at√© que a diferen√ßa de $x_i$ para $x_{i+1}$ e de $f(x_i)$ e $0$ sejam menores que os seus respectivos erros, temos que:

$x_{i+1} = x_i - \frac{f(x_i)}{f'(x_i)}$, at√© que $\lvert x_{i+1} - x_{i} \rvert < \varepsilon_y$ ou $\lvert f(x_i) \rvert < \varepsilon_y$, para qualquer $f$.

Mas vamos assumir que $g(x) = - f(x) \Rightarrow g'(x) = -f'(x)$ e, como a formula de Aproxima√ß√£o de Newton pode ser usada para qualquer $f$, temos que, para $g$:

$x_{i+1} = x_i - \frac{g(x_i)}{g'(x_i)} = x_i - \frac{-f(x_i)}{-f'(x_i)} = x_i - \frac{(-1) \cdot f(x_i)}{(-1) \cdot f'(x_i)} =  x_i - \frac{f(x_i)}{f'(x_i)}$

Que √© exatamente as f√≥rmulas de itera√ß√£o de Newton para $f$, provando o argumento.

Se supormos que a derivada de $f$ na sua raiz √© positiva (e $f'(x)$ nunca √© zero), isso significa que a fun√ß√£o, a partir de $z$ (a raiz) possui inclina√ß√£o sempre positiva, e que as aproxima√ß√µes da sequ√™ncia $x_0, ..., x_n$ sempre existem (lembrando que se $f'(z) > 0 \Rightarrow g'(z) < 0$ e os mesmos argumentos valem, tamb√©m, para $g$.

### Primeiro Caso (polinomial)

Para esse caso, temos:

$f(x) = \frac{1}{2}\left( x^2 -1 \right) \Rightarrow f(1) = 0$

$f'(x) = x \Rightarrow f'(1) = 1$

$f''(x) = 1 \Rightarrow f''(1) = 1 > 0$

```python
# Dois mil pontos entre 0 e 2:
interval = np.arange(0, 2, 0.001)

# Functions:
def f(x):
    return (1/2)*(np.power(x,2)-1)

def df(x):
    return x

def ddf(x):
    return np.array([1 for i in x])

fig, ax = plt.subplots()

# Plots:
plt.plot(interval, f(interval), label="f(x) Polinomial", linewidth=3)
plt.plot(interval, df(interval), label="f'(x) Polinomial", linewidth=3)
plt.plot(interval, ddf(interval), label="f''(x) Polinomial", linewidth=3)

# Grid:
ax.set_axisbelow(True)
ax.minorticks_on()
ax.grid(which='major', linestyle='-', linewidth='0.5', color='red')
ax.grid(which='minor', linestyle=':', linewidth='0.5', color='black')

# Legend:
plt.legend()
```

## Quest√£o 1.4 - Resposta:

Devemos entender que para o m√©todo de Newton ser invariante para transla√ß√£o, **todos os pontos da aproxima√ß√£o, inclusive o ponto inicial $x_0$** deve sofrer a mesma transla√ß√£o.

Assim, considerando a Aproxima√ß√£o de Newton para $f$, para uma vari√°vel $\bar{x}$ √© dada por:

$\bar{x}_{i+1} = \bar{x}_i - \frac{f(\bar{x}_i)}{f'(\bar{x}_i)}$

Se assumimos que $g$ √© uma transla√ß√£o de $f$, e aplicamos essa transla√ß√£o em todos os $\bar{x_i}$, tem-se:

$\bar{x}_i = x_i + b, \forall i \Rightarrow \bar{x}_{i+1} = x_{i+1} + b$, pela recorr√™ncia do M√©todo de Newton:

$\bar{x}_{i+1} = x_i + b - \frac{f(x_i + b)}{f'(x_i + b)} = x_i - \frac{f(x_i + b)}{f'(x_i + b)} + b = x_i - \frac{g(x_i)}{g'(x_i)} + b = x_{i+1} + b$

Portanto:

$x_{i+1} + b = x_i + b - \frac{f(x_i + b)}{f'(x_i + b)} $

Mas lembrem-se que assumimos que $\bar{x} = x + b$, ent√£o, se tomamos $x = \bar{x}$:

$\bar{x}_{i+1} + b = \bar{x}_i + b - \frac{f(\bar{x}_i + b)}{f'(\bar{x}_i + b)} \Leftrightarrow x_{i+1} = x_i - \frac{g(x_i)}{g'(x_i)}$

Mostrando que todos os pontos de aproxima√ß√£o do M√©todo de Newton s√£o transladados por $b$, se $g$ for uma transla√ß√£o por $b$ de $f$ e o ponto inicial da aproxima√ß√£o de $g$ sofrer uma transla√ß√£o $b$, em rela√ß√£o ao ponto inicial da aproxima√ß√£o de $f$.

## Quest√£o 1.5 - Resposta:

Vamos aproximar por Newton uma fun√ß√£o bastante simples:

$f(x) = (x-1)^2$

$g(x) = f(x+1) = x^2$

```python
# A fun√ß√£o f:
def f(x):
    return np.power(x - 1, 2)

# A fun√ß√£o f':
def df(x):
    return 2 * (x - 1)

# A fun√ß√£o g (f transladada em 1):
def g(x):
    return f(x+1)

# A fun√ß√£o g' = f'(x+1) (f' transladada em 1)
def dg(x):
    return df(x+1)

# M√©todo de Newton recuperado de um dos testes:
def newton_pts(f, df, x, xtol=1e-8, ytol=1e-8, maxiter=100):
    """
    M√©todo de Newton para encontrar uma raiz de  f  a partir do valor inicial  x.
    Retorna todos os pontos pelos quais o algoritmo passa.
    
    O algoritmo termina quando o tamanho do passo √© menor do que  xtol,
    quando o valor absoluto da fun√ß√£o ficar abaixo de  ytol,
    ou ap√≥s  maxiter  itera√ß√µes.
    """
    res = [x]
    while len(res) <= maxiter:
        nx = res[-1] - f(res[-1])/df(res[-1])
        res.append(nx)
        if np.abs(f(nx)) < ytol or np.abs(res[-1] - res[-2]) < xtol:
            return res
    return res
```

```python
xxs = np.arange(-2, 3, 0.01)

fig, ax = plt.subplots()

# Plots:
plt.plot(xxs, f(xxs), label="f", linewidth=3)
plt.plot(xxs, g(xxs), label="g", linewidth=3)

# Grid:
ax.set_axisbelow(True)
ax.minorticks_on()
ax.grid(which='major', linestyle='-', linewidth='0.5', color='red')
ax.grid(which='minor', linestyle=':', linewidth='0.5', color='black')

plt.legend()
```

```python
# Pontos originais da busca dos zeros:
x = 5
x_bar = x + 1 # x_bar √© x transladada

x_bar_points = newton_pts(f, df, x_bar)
x_points = newton_pts(g, dg, x)
difference = np.array(x_bar_points) - np.array(x_points)

fig, ax = plt.subplots()

# Plots:
plt.plot(x_bar_points, label="Pontos da aproxima√ß√£o de f")
plt.plot(x_points, label="Pontos da aproxima√ß√£o de g")
plt.plot(difference, label="Diferen√ßa dos pontos obtidos para f vs. g")

# Grid:
ax.set_axisbelow(True)
ax.minorticks_on()
ax.grid(which='major', linestyle='-', linewidth='0.25', color='red')
ax.grid(which='minor', linestyle=':', linewidth='0.15', color='blue')

plt.legend()
```

## Quest√£o 1.8 - Resposta:

√â considerada a transforma√ß√£o de $f$ em $g$, pela seguinte rela√ß√£o:

$g(x) = Cf(ax+b)$, $f(z) = 0$, $f''(z) \neq 0$

E que $g(0) = 0$, $g'(0) = 1$ e $g''(0) = 1$.

**Observa√ß√£o:** Devemos entender que, para atender os crit√©rios acima, tamb√©m $f'(z) \neq 0$, porque $g'(0)$ ser√° um m√∫ltiplo de $f'(z)$

Vamos para o c√°lculo de $a$, $b$ e $C$ propriamente ditos:


$g(0) = Cf(a \cdot 0 + b) = Cf(z) = C \cdot 0 = 0 \Rightarrow a \cdot 0 + b = z \Rightarrow b = z$

$g'(x) = Cf'(ax + b) \cdot a = Caf'(ax + b) = Caf'(ax + z) \Rightarrow g'(0) = Caf'(z) = 1 \Rightarrow Ca = \frac{1}{f'(z)}$, requerendo $f'(z) \neq 0$

$g''(x) = Caf''(ax+z) \cdot a = Ca^2f''(ax + z) \Rightarrow g''(0) = Ca^2f''(z) = 1 \Rightarrow Ca^2 = \frac{1}{f''(z)}$

$ \Rightarrow \frac{Ca}{Ca^2} = \frac{\frac{1}{f'(z)}}{\frac{1}{f''(z)}} = \frac{f''(z)}{f'(z)} \Rightarrow \frac{1}{a} = \frac{f''(z)}{f'(z)} \Rightarrow a = \frac{f'(z)}{f''(z)}$, $f'(z) \neq 0$ e $f''(z) \neq 0$

$ \Rightarrow Ca = \frac{1}{f'(z)} \Rightarrow C \cdot \frac{f'(z)}{f''(z)} = \frac{1}{f'(z)} \Rightarrow C = \frac{f''(z)}{\left[ f'(z) \right]^2}$

Assim, para $g(0) = 0$, $g'(0) = 1$ e $g''(0) = 1$, devemos ter:

* $a = \frac{f'(z)}{f''(z)}$
* $b = z$
* $C = \frac{f''(z)}{\left[ f'(z) \right]^2}$

## Quest√£o 1.9 - Resposta:

√à poss√≠vel ver que todos os pontos da aproxima√ß√£o de $g$ s√£o transladados e escalados por $b$ e $a$, respectivamente, de forma que se $(x_i)_{i \in \mathbb{N}}$ √© a sequ√™ncia de pontos de aproxima√ß√£o de $g$, os pontos de aproxima√ß√£o de $f$ s√£o dados por $(\bar{x}_i = ax_i + b)_{i \in \mathbb{N}}$

Demonstra√ß√£o:

Seja $g(x) = Cf(ax + z)$ e $\bar{x} = ax + z$

A sequ√™ncia em $i \in \mathbb{N} $ das aproxima√ß√µes do zero de $f$ √© dada por:

$x_{i+1} = x_i - \frac{f(x_i)}{f'(x_i)}$

E a sequencia das aproxima√ß√µes do zero de $g$ √© dada por:

$x_{i+1} = x_i - \frac{g(x_i)}{g'(x_i)}$

E podemos desenvolver a aproxima√ß√£o de $g$ e chegar na aproxima√ß√£o de $f$, assim:

$x_{i+1} =  x_i - \frac{g(x_i)}{g'(x_i)} =  x_i - \frac{Cf(ax_i + z)}{ \frac{d}{dx}\left[Cf(ax_i + z)\right] } = x_i - \frac{Cf(ax_i + z)}{ Caf'(ax_i + z) } = x_i - \frac{f(ax_i + z)}{ af'(ax_i + z) }$

$\Rightarrow ax_{i+1} = ax_i - \frac{f(ax_i + z)}{ f'(ax_i + z) } \Rightarrow ax_{i+1} + z = ax_i - \frac{f(ax_i + z)}{ f'(ax_i + z) } + z => \bar{x}_{i+1} = ax_i - \frac{f(\bar{x_i})}{f'(\bar{x_i})} + z = ax_i + z - \frac{f(\bar{x_i})}{f'(\bar{x_i})} = \bar{a}x_i - \frac{f(\bar{x_i})}{f'(\bar{x_i})} $

Assim, todos os pontos da aproxima√ß√£o de $f$, em rela√ß√£o aos pontos de $g$, est√£o sofrendo a transforma√ß√£o $T(x) = ax + z$, enquanto o fator de escala $C$ √© irrelevante para a sequ√™ncia da Aproxima√ß√£o de Newton.

```python
# Exemplo do descrito acima:

# A fun√ß√£o f:
def f(x):
    return np.power(x - 1, 2)

# A fun√ß√£o f':
def df(x):
    return 2 * (x - 1)

# A fun√ß√£o g (f transladada e escalada em x e em y):
def g(x):
    return 0.5*f(1.5*x+3)

# A fun√ß√£o g' = f'(x+1) (f' transladada e escalada em x e em y)
def dg(x):
    return 0.5*1.5*df(1.5*x+3)
```

```python
xxs = np.arange(-2, 2, 0.01)

fig, ax = plt.subplots()

# Plots:
plt.plot(xxs, f(xxs), label="f", linewidth=3)
plt.plot(xxs, g(xxs), label="g", linewidth=3)

# Grid:
ax.set_axisbelow(True)
ax.minorticks_on()
ax.grid(which='major', linestyle='-', linewidth='0.5', color='red')
ax.grid(which='minor', linestyle=':', linewidth='0.5', color='black')

plt.legend()
```

```python
# Pontos originais da busca dos zeros:
x = 5
x_bar = 1.5*x + 3 # x_bar √© x transformada

x_bar_points = newton_pts(f, df, x_bar)
x_points = newton_pts(g, dg, x)
evaluated_x_bar_points = np.array(x_points) * 1.5 + 3
difference = evaluated_x_bar_points - x_bar_points

fig, ax = plt.subplots()

# Plots:
plt.plot(x_bar_points, label="Pontos da aproxima√ß√£o de f")
plt.plot(x_points, label="Pontos da aproxima√ß√£o de g")
plt.plot(difference, label="Pontos de f vs. T dos pontos de g")

# Grid:
ax.set_axisbelow(True)
ax.minorticks_on()
ax.grid(which='major', linestyle='-', linewidth='0.25', color='red')
ax.grid(which='minor', linestyle=':', linewidth='0.15', color='blue')

plt.legend()
```

## Quest√£o 10 - Resposta:

Para o m√©todo da Secante e para o m√©todo da Bisse√ß√£o, tanto as transforma√ß√µes de escala e as transla√ß√µes s√£o irrelevantes para a efici√™ncia do m√©todo. Os pontos encontrados em uma fun√ß√£o transladada e escalada s√£o os pontos da fun√ß√£o original, aplicadas as transla√ß√µes e as escalas. Se o escalonamento for feito na vari√°vel dependente, o efeito sobre a aproxima√ß√£o n√£o existe.

√â neces≈õaria uma demonstra√ß√£o para o M√©todo da Secante e outra demonstra√ß√£o para o M√©todo da Bisse√ß√£o.

Primeiro assumimos que:

$g(x) = Cf(ax + z)$
