## Questão 1 .1 - Resposta:

Podemos considerar uma função $f$, e uma tangente a ela, conforme a figura abaixo:


Pela figura, vemos que a tangente passa por $f$ no ponto $(x_0, f(x_0) = y_0)$ e que o ponto $x_1$ pode ser encontrado por:

$$
\frac{\Delta y}{\Delta x} = \frac{y_0 - 0}{x_0 - x_1} = f'(x_0) \Rightarrow x_1 = x_0 - \frac{f(x_0)}{f'(x_0)}
$$

Como temos que aplicar reiteiradamente o método, até que a diferença de $x_i$ para $x_{i+1}$ e de $f(x_i)$ e $0$ sejam menores que os seus respectivos erros, temos que:

$x_{i+1} = x_i - \frac{f(x_i)}{f'(x_i)}$, até que $\lvert x_{i+1} - x_{i} \rvert < \varepsilon_y$ ou $\lvert f(x_i) \rvert < \varepsilon_y$, para qualquer $f$.

Mas vamos assumir que $g(x) = - f(x) \Rightarrow g'(x) = -f'(x)$ e, como a formula de Aproximação de Newton pode ser usada para qualquer $f$, temos que, para $g$:

$x_{i+1} = x_i - \frac{g(x_i)}{g'(x_i)} = x_i - \frac{-f(x_i)}{-f'(x_i)} = x_i - \frac{(-1) \cdot f(x_i)}{(-1) \cdot f'(x_i)} =  x_i - \frac{f(x_i)}{f'(x_i)}$

Que é exatamente as fórmulas de iteração de Newton para $f$, provando o argumento.

Se supormos que a derivada de $f$ na sua raiz é positiva (e $f'(x)$ nunca é zero), isso significa que a função, a partir de $z$ (a raiz) possui inclinação sempre positiva, e que as aproximações da sequência $x_0, ..., x_n$ sempre existem (lembrando que se $f'(z) > 0 \Rightarrow g'(z) < 0$ e os mesmos argumentos valem, também, para $g$.

## Questão 1.4 - Resposta:

Devemos entender que para o método de Newton ser invariante para translação, **todos os pontos da aproximação, inclusive o ponto inicial $x_0$** deve sofrer a mesma translação.

Assim, considerando a Aproximação de Newton para $f$, para uma variável $\bar{x}$ é dada por:

$\bar{x}_{i+1} = \bar{x}_i - \frac{f(\bar{x}_i)}{f'(\bar{x}_i)}$

Se assumimos que $g$ é uma translação de $f$, e aplicamos essa translação em todos os $\bar{x_i}$, tem-se:

$\bar{x}_i = x_i + b, \forall i \Rightarrow \bar{x}_{i+1} = x_{i+1} + b$, pela recorrência do Método de Newton:

$\bar{x}_{i+1} = x_i + b - \frac{f(x_i + b)}{f'(x_i + b)} = x_i - \frac{f(x_i + b)}{f'(x_i + b)} + b = x_i - \frac{g(x_i)}{g'(x_i)} + b = x_{i+1} + b$

Portanto:

$x_{i+1} + b = x_i + b - \frac{f(x_i + b)}{f'(x_i + b)} $

Mas lembrem-se que assumimos que $\bar{x} = x + b$, então, se tomamos $x = \bar{x}$:

$\bar{x}_{i+1} + b = \bar{x}_i + b - \frac{f(\bar{x}_i + b)}{f'(\bar{x}_i + b)} \Leftrightarrow x_{i+1} = x_i - \frac{g(x_i)}{g'(x_i)}$

Mostrando que todos os pontos de aproximação do Método de Newton são transladados por $b$, se $g$ for uma translação por $b$ de $f$ e o ponto inicial da aproximação de $g$ sofrer uma translação $b$, em relação ao ponto inicial da aproximação de $f$.

## Questão 1.5 - Resposta:

Vamos aproximar por Newton uma função bastante simples:

$f(x) = (x-1)^2$

$g(x) = f(x+1) = x^2$

```python
# A função f:
def f(x):
    return np.power(x - 1, 2)

# A função f':
def df(x):
    return 2 * (x - 1)

# A função g (f transladada em 1):
def g(x):
    return f(x+1)

# A função g' = f'(x+1) (f' transladada em 1)
def dg(x):
    return df(x+1)

# Método de Newton recuperado de um dos testes:
def newton_pts(f, df, x, xtol=1e-8, ytol=1e-8, maxiter=100):
    """
    Método de Newton para encontrar uma raiz de  f  a partir do valor inicial  x.
    Retorna todos os pontos pelos quais o algoritmo passa.
    
    O algoritmo termina quando o tamanho do passo é menor do que  xtol,
    quando o valor absoluto da função ficar abaixo de  ytol,
    ou após  maxiter  iterações.
    """
    res = [x]
    while len(res) <= maxiter:
        nx = res[-1] - f(res[-1])/df(res[-1])
        res.append(nx)
        if np.abs(f(nx)) < ytol or np.abs(res[-1] - res[-2]) < xtol:
            return res
    return res
```

```python
xxs = np.arange(-2, 3, 0.01)

fig, ax = plt.subplots()

# Plots:
plt.plot(xxs, f(xxs), label="f", linewidth=3)
plt.plot(xxs, g(xxs), label="g", linewidth=3)

# Grid:
ax.set_axisbelow(True)
ax.minorticks_on()
ax.grid(which='major', linestyle='-', linewidth='0.5', color='red')
ax.grid(which='minor', linestyle=':', linewidth='0.5', color='black')

plt.legend()
```

```python
# Pontos originais da busca dos zeros:
x = 5
x_bar = x + 1 # x_bar é x transladada

x_bar_points = newton_pts(f, df, x_bar)
x_points = newton_pts(g, dg, x)
difference = np.array(x_bar_points) - np.array(x_points)

fig, ax = plt.subplots()

# Plots:
plt.plot(x_bar_points, label="Pontos da aproximação de f")
plt.plot(x_points, label="Pontos da aproximação de g")
plt.plot(difference, label="Diferença dos pontos obtidos para f vs. g")

# Grid:
ax.set_axisbelow(True)
ax.minorticks_on()
ax.grid(which='major', linestyle='-', linewidth='0.25', color='red')
ax.grid(which='minor', linestyle=':', linewidth='0.15', color='blue')

plt.legend()
```

## Questão 1.8 - Resposta:

É considerada a transformação de $f$ em $g$, pela seguinte relação:

$g(x) = Cf(ax+b)$, $f(z) = 0$, $f''(z) \neq 0$

E que $g(0) = 0$, $g'(0) = 1$ e $g''(0) = 1$.

**Observação:** Devemos entender que, para atender os critérios acima, também $f'(z) \neq 0$, porque $g'(0)$ será um múltiplo de $f'(z)$

Vamos para o cálculo de $a$, $b$ e $C$ propriamente ditos:


$g(0) = Cf(a \cdot 0 + b) = Cf(z) = C \cdot 0 = 0 \Rightarrow a \cdot 0 + b = z \Rightarrow b = z$

$g'(x) = Cf'(ax + b) \cdot a = Caf'(ax + b) = Caf'(ax + z) \Rightarrow g'(0) = Caf'(z) = 1 \Rightarrow Ca = \frac{1}{f'(z)}$, requerendo $f'(z) \neq 0$

$g''(x) = Caf''(ax+z) \cdot a = Ca^2f''(ax + z) \Rightarrow g''(0) = Ca^2f''(z) = 1 \Rightarrow Ca^2 = \frac{1}{f''(z)}$

$ \Rightarrow \frac{Ca}{Ca^2} = \frac{\frac{1}{f'(z)}}{\frac{1}{f''(z)}} = \frac{f''(z)}{f'(z)} \Rightarrow \frac{1}{a} = \frac{f''(z)}{f'(z)} \Rightarrow a = \frac{f'(z)}{f''(z)}$, $f'(z) \neq 0$ e $f''(z) \neq 0$

$ \Rightarrow Ca = \frac{1}{f'(z)} \Rightarrow C \cdot \frac{f'(z)}{f''(z)} = \frac{1}{f'(z)} \Rightarrow C = \frac{f''(z)}{\left[ f'(z) \right]^2}$

Assim, para $g(0) = 0$, $g'(0) = 1$ e $g''(0) = 1$, devemos ter:

* $a = \frac{f'(z)}{f''(z)}$
* $b = z$
* $C = \frac{f''(z)}{\left[ f'(z) \right]^2}$

## Questão 1.9 - Resposta:

È possível ver que todos os pontos da aproximação de $g$ são transladados e escalados por $b$ e $a$, respectivamente, de forma que se $(x_i)_{i \in \mathbb{N}}$ é a sequência de pontos de aproximação de $g$, os pontos de aproximação de $f$ são dados por $(\bar{x}_i = ax_i + b)_{i \in \mathbb{N}}$

Demonstração:

Seja $g(x) = Cf(ax + z)$ e $\bar{x} = ax + z$

A sequência em $i \in \mathbb{N} $ das aproximações do zero de $f$ é dada por:

$x_{i+1} = x_i - \frac{f(x_i)}{f'(x_i)}$

E a sequencia das aproximações do zero de $g$ é dada por:

$x_{i+1} = x_i - \frac{g(x_i)}{g'(x_i)}$

E podemos desenvolver a aproximação de $g$ e chegar na aproximação de $f$, assim:

$x_{i+1} =  x_i - \frac{g(x_i)}{g'(x_i)} =  x_i - \frac{Cf(ax_i + z)}{ \frac{d}{dx}\left[Cf(ax_i + z)\right] } = x_i - \frac{Cf(ax_i + z)}{ Caf'(ax_i + z) } = x_i - \frac{f(ax_i + z)}{ af'(ax_i + z) }$

$\Rightarrow ax_{i+1} = ax_i - \frac{f(ax_i + z)}{ f'(ax_i + z) } \Rightarrow ax_{i+1} + z = ax_i - \frac{f(ax_i + z)}{ f'(ax_i + z) } + z => \bar{x}_{i+1} = ax_i - \frac{f(\bar{x_i})}{f'(\bar{x_i})} + z = ax_i + z - \frac{f(\bar{x_i})}{f'(\bar{x_i})} = \bar{a}x_i - \frac{f(\bar{x_i})}{f'(\bar{x_i})} $

Assim, todos os pontos da aproximação de $f$, em relação aos pontos de $g$, estão sofrendo a transformação $T(x) = ax + z$, enquanto o fator de escala $C$ é irrelevante para a sequência da Aproximação de Newton.

```python
# Exemplo do descrito acima:

# A função f:
def f(x):
    return np.power(x - 1, 2)

# A função f':
def df(x):
    return 2 * (x - 1)

# A função g (f transladada e escalada em x e em y):
def g(x):
    return 0.5*f(1.5*x+3)

# A função g' = f'(x+1) (f' transladada e escalada em x e em y)
def dg(x):
    return 0.5*1.5*df(1.5*x+3)
```

```python
xxs = np.arange(-2, 2, 0.01)

fig, ax = plt.subplots()

# Plots:
plt.plot(xxs, f(xxs), label="f", linewidth=3)
plt.plot(xxs, g(xxs), label="g", linewidth=3)

# Grid:
ax.set_axisbelow(True)
ax.minorticks_on()
ax.grid(which='major', linestyle='-', linewidth='0.5', color='red')
ax.grid(which='minor', linestyle=':', linewidth='0.5', color='black')

plt.legend()
```

```python
# Pontos originais da busca dos zeros:
x = 5
x_bar = 1.5*x + 3 # x_bar é x transformada

x_bar_points = newton_pts(f, df, x_bar)
x_points = newton_pts(g, dg, x)
evaluated_x_bar_points = np.array(x_points) * 1.5 + 3
difference = evaluated_x_bar_points - x_bar_points

fig, ax = plt.subplots()

# Plots:
plt.plot(x_bar_points, label="Pontos da aproximação de f")
plt.plot(x_points, label="Pontos da aproximação de g")
plt.plot(difference, label="Pontos de f vs. T dos pontos de g")

# Grid:
ax.set_axisbelow(True)
ax.minorticks_on()
ax.grid(which='major', linestyle='-', linewidth='0.25', color='red')
ax.grid(which='minor', linestyle=':', linewidth='0.15', color='blue')

plt.legend()
```

## Questão 10 - Resposta:

Para o método da Secante e para o método da Bisseção, tanto as transformações de escala e as translações são irrelevantes para a eficiência do método. Os pontos encontrados em uma função transladada e escalada são os pontos da função original, aplicadas as translações e as escalas. Se o escalonamento for feito na variável dependente, o efeito sobre a aproximação não existe.

É necesśaria uma demonstração para o Método da Secante e outra demonstração para o Método da Bisseção.

Primeiro assumimos que:

$g(x) = Cf(ax + z)$ 

E assumimos a seguinte transformação: 

$\bar x = ax + z$

### Método da Secante

Podemos dizer, de acordo com o método da secante, que a seguinte igualdade vale:

$\bar x_{i+2} \equiv \bar x_i - \frac{f(\bar x_i)}{ \frac{f(\bar x_{i+1}) - f(\bar x_i)}{\bar x_{i+1} - x_i} } \Rightarrow \bar x_{i+2} = \frac{\bar x_i f(\bar x_{i+1}) - \bar x_{i_1} f(\bar x_i)}{ f(\bar x_{i+1}) - f(\bar x_i)}$

Aplicando diretamente a definição em $g$, temos a seguinte recorrência, assumindo:

Dados:
* $\bar x_0 = ax_0 + z$
* $\bar x_1 = ax_1 + z$
* $\bar x_i = ax_i + z$

$ \Rightarrow x_{i+2} = x_i - \frac{g(x_i)}{ \frac{g(x_{i+1}) - g(x_i)}{x_{i+1} - x_i} } 
= x_i - \frac{Cf(ax_i + z)}{ \frac{Cf(ax_{i+1}+z) - Cf(ax_i + z)}{x_{i+1} - x_i} } 
= x_i - \frac{Cf(ax_i + z)}{ C \frac{f(ax_{i+1}+z) - f(ax_i + z)}{x_{i+1} - x_i} }
= x_i - \frac{f(\bar x_i)}{ \frac{f(\bar x_{i+1}) - f(\bar x_i)}{x_{i+1} - x_i} }$

$\Rightarrow ax_{i+2} 
= a \left( x_i - \frac{f(\bar x_i)}{ \frac{f(\bar x_{i+1}) - f(\bar x_i)}{x_{i+1} - x_i} } \right) 
= ax_i - \frac{f(\bar x_i)}{ \frac{1}{a} \frac{f(\bar x_{i+1}) - f(\bar x_i)}{x_{i+1} - x_i} }
= ax_i - \frac{f(\bar x_i)}{ \frac{f(\bar x_{i+1}) - f(\bar x_i)}{ax_{i+1} - ax_i} } 
= ax_i - \frac{f(\bar x_i)}{ \frac{f(\bar x_{i+1}) - f(\bar x_i)}{ax_{i+1} + z - ax_i -z } } 
= ax_i - \frac{f(\bar x_i)}{ \frac{f(\bar x_{i+1}) - f(\bar x_i)}{\bar x_{i+1} - \bar x_i } }$

$\Rightarrow ax_{i+2} + z
= ax_i + z - \frac{f(\bar x_i)}{ \frac{f(\bar x_{i+1}) - f(\bar x_i)}{\bar x_{i+1} + \bar x_i } }
\Rightarrow \bar x_{i+2} = \bar x_i - \frac{f(\bar x_i)}{ \frac{f(\bar x_{i+1}) - f(\bar x_i)}{\bar x_{i+1} + \bar x_i } }$

Que é exatamente a aproximação de f, com o Método da Secante, com todos os pontos de G transformados por $4T(x) = \bar x = ax + z$, como queríamos demonstrar.

### Método da Bisseção

Para o Método da Bisseção, precisamos definir dois pontos iniciais, e o início do processo de recorrência.

Assim, temos a sequência inicial de pontos dada por:

$x_0, x_1, x_2 = \frac{x_0 + x_1}{2}$, dados $x_0 = a$ e $x_1 = b$, $a < b, a,b \in \mathbb{R}$

Assim, a regra de recorrência dada a partir de i = 3, é:

$
x_{i+3} = 
\begin{cases}
      x_i & \text{se } f(x_i) = 0\\
      x_{i+1} & \text{se } f(x_{i+1})  = 0\\
      x_{i+2} & \text{se } f(x_{i+2})  = 0\\
      \frac{x_{i+1} + x_{i+2}}{2} & \text{se } f(x_i) \cdot f(x_{i+2}) > 0\\
      \frac{x_i + x_{i+2}}{2} & \text{se } f(x_i) \cdot f(x_{i+2}) < 0\\
    \end{cases}       
$

Agora vamos aplicar a mesma regra de bisseção para a função $g(x) = Cf(ax + z)$, com $\bar x = ax + z$:

$
x_{i+3} = 
\begin{cases}
      x_i & \text{se } g(x_i) = 0\\
      x_{i+1} & \text{se } g(x_{i+1})  = 0\\
      x_{i+2} & \text{se } g(x_{i+2})  = 0\\
      \frac{x_{i+1} + x_{i+2}}{2} & \text{se } g(x_i) \cdot g(x_{i+2}) > 0\\
      \frac{x_i + x_{i+2}}{2} & \text{se } g(x_i) \cdot g(x_{i+2}) < 0\\
    \end{cases}
= 
\begin{cases}
      x_i & \text{se } Cf(ax_i + z) = 0\\
      x_{i+1} & \text{se } Cf(ax_{i+1} + z)  = 0\\
      x_{i+2} & \text{se } Cf(ax_{i+2} + z)  = 0\\
      \frac{x_{i+1} + x_{i+2}}{2} & \text{se } Cf(ax_i + z) \cdot Cf(ax_{i+2} + z) > 0\\
      \frac{x_i + x_{i+2}}{2} & \text{se } Cf(ax_i + z) \cdot Cf(ax_{i+2} + z) < 0\\
    \end{cases}
= 
\begin{cases}
      x_i & \text{se } Cf(ax_i + z) = 0\\
      x_{i+1} & \text{se } Cf(ax_{i+1} + z)  = 0\\
      x_{i+2} & \text{se } Cf(ax_{i+2} + z)  = 0\\
      \frac{x_{i+1} + x_{i+2}}{2} & \text{se } C^2f(ax_i + z) \cdot f(ax_{i+2} + z) > 0\\
      \frac{x_i + x_{i+2}}{2} & \text{se } C^2f(ax_i + z) \cdot f(ax_{i+2} + z) < 0\\
    \end{cases}
= 
\begin{cases}
      x_i & \text{se } f(ax_i + z) = 0, \text{pois } C \neq 0\\
      x_{i+1} & \text{se } f(ax_{i+1} + z)  = 0, \text{pois } C \neq 0\\
      x_{i+2} & \text{se } f(ax_{i+2} + z)  = 0, \text{pois } C \neq 0\\
      \frac{x_{i+1} + x_{i+2}}{2} & \text{se } f(ax_i + z) \cdot f(ax_{i+2} + z) > 0, \text{pois } C^2 > 0\\
      \frac{x_i + x_{i+2}}{2} & \text{se } f(ax_i + z) \cdot f(ax_{i+2} + z) < 0, \text{pois } C^2 > 0\\
    \end{cases}
$

$ 
\Rightarrow ax_{i+3} =  \begin{cases}
      ax_i & \text{se } f(ax_i + z) = 0\\
      ax_{i+1} & \text{se } f(ax_{i+1} + z)  = 0\\
      ax_{i+2} & \text{se } f(ax_{i+2} + z)  = 0\\
      a\frac{x_{i+1} + x_{i+2}}{2} & \text{se } f(ax_i + z) \cdot f(ax_{i+2} + z) > 0\\
      a\frac{x_i + x_{i+2}}{2} & \text{se } f(ax_i + z) \cdot f(ax_{i+2} + z) < 0\\
    \end{cases}
$

$ 
\Rightarrow ax_{i+3} + z =  \begin{cases}
      ax_i + z  & \text{se } f(ax_i + z) = 0\\
      ax_{i+1} + z  & \text{se } f(ax_{i+1} + z)  = 0\\
      ax_{i+2} + z  & \text{se } f(ax_{i+2} + z)  = 0\\
      a\frac{x_{i+1} + x_{i+2}}{2} + z  & \text{se } f(ax_i + z) \cdot f(ax_{i+2} + z) > 0\\
      a\frac{x_i + x_{i+2}}{2} + z  & \text{se } f(ax_i + z) \cdot f(ax_{i+2} + z) < 0\\
    \end{cases}
=  
\begin{cases}
      ax_i + z  & \text{se } f(ax_i + z) = 0\\
      ax_{i+1} + z  & \text{se } f(ax_{i+1} + z)  = 0\\
      ax_{i+2} + z  & \text{se } f(ax_{i+2} + z)  = 0\\
      \frac{ax_{i+1} + z + ax_{i+2} + z}{2}  & \text{se } f(ax_i + z) \cdot f(ax_{i+2} + z) > 0\\
      \frac{ax_i + z + ax_{i+2} + z}{2}  & \text{se } f(ax_i + z) \cdot f(ax_{i+2} + z) < 0\\
    \end{cases}
$

$ 
\Rightarrow \bar x_{i+3} =  \begin{cases}
      \bar x_i  & \text{se } f(\bar x_i) = 0\\
      \bar x_{i+1}  & \text{se } f(\bar x_{i+1})  = 0\\
      \bar x_{i+2}  & \text{se } f(\bar x_{i+2})  = 0\\
      \frac{\bar x_{i+1} + \bar x_{i+2}}{2}  & \text{se } f(\bar x_i) \cdot f(\bar x_{i+2}) > 0\\
      \frac{\bar x_i + \bar x_{i+2}}{2}  & \text{se } f(\bar x_i) \cdot f(\bar x_{i+2}) < 0\\
    \end{cases}
$,

que é exatamente a aproximação por bisseção para $f$, com todos os pontos (incluindo os iniciais), transformados por $T(x) = ax + z$, como queríamos demonstrar.
