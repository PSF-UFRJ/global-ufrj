## Questão 1 .1 - Resposta:

Podemos considerar uma função $f$, e uma tangente a ela, conforme a figura abaixo:


Pela figura, vemos que a tangente passa por $f$ no ponto $(x_0, f(x_0) = y_0)$ e que o ponto $x_1$ pode ser encontrado por:

$$
\frac{\Delta y}{\Delta x} = \frac{y_0 - 0}{x_0 - x_1} = f'(x_0) \Rightarrow x_1 = x_0 - \frac{f(x_0)}{f'(x_0)}
$$

Como temos que aplicar reiteiradamente o método, até que a diferença de $x_i$ para $x_{i+1}$ e de $f(x_i)$ e $0$ sejam menores que os seus respectivos erros, temos que:

$x_{i+1} = x_i - \frac{f(x_i)}{f'(x_i)}$, até que $\lvert x_{i+1} - x_{i} \rvert < \varepsilon_y$ ou $\lvert f(x_i) \rvert < \varepsilon_y$, para qualquer $f$.

Mas vamos assumir que $g(x) = - f(x) \Rightarrow g'(x) = -f'(x)$ e, como a formula de Aproximação de Newton pode ser usada para qualquer $f$, temos que, para $g$:

$x_{i+1} = x_i - \frac{g(x_i)}{g'(x_i)} = x_i - \frac{-f(x_i)}{-f'(x_i)} = x_i - \frac{(-1) \cdot f(x_i)}{(-1) \cdot f'(x_i)} =  x_i - \frac{f(x_i)}{f'(x_i)}$

Que é exatamente as fórmulas de iteração de Newton para $f$, provando o argumento.

Se supormos que a derivada de $f$ na sua raiz é positiva (e $f'(x)$ nunca é zero), isso significa que a função, a partir de $z$ (a raiz) possui inclinação sempre positiva, e que as aproximações da sequência $x_0, ..., x_n$ sempre existem (lembrando que se $f'(z) > 0 \Rightarrow g'(z) < 0$ e os mesmos argumentos valem, também, para $g$.

Para responder essa questão, devemos, primeiro, encontrar três funções que atendam a esses critérios e, então, esboçá-los.

```python
# Dependências:
import numpy as np
import matplotlib.pyplot as plt
```

### Primeiro Caso (polinomial)

Para esse caso, temos:

$f(x) = \frac{1}{2}\left( x^2 -1 \right) \Rightarrow f(1) = 0$

$f'(x) = x \Rightarrow f'(1) = 1$

$f''(x) = 1 \Rightarrow f''(1) = 1 > 0$

```python
# Dois mil pontos entre 0 e 2:
interval = np.arange(0, 2, 0.001)

# Functions:
def f(x):
    return (1/2)*(np.power(x,2)-1)

def df(x):
    return x

def ddf(x):
    return np.array([1 for i in x])

fig, ax = plt.subplots()

# Plots:
plt.plot(interval, f(interval), label="f(x) Polinomial", linewidth=3)
plt.plot(interval, df(interval), label="f'(x) Polinomial", linewidth=3)
plt.plot(interval, ddf(interval), label="f''(x) Polinomial", linewidth=3)

# Grid:
ax.set_axisbelow(True)
ax.minorticks_on()
ax.grid(which='major', linestyle='-', linewidth='0.5', color='red')
ax.grid(which='minor', linestyle=':', linewidth='0.5', color='black')

# Legend:
plt.legend()
```

### Segundo Caso (polinomial)

Para esse caso, temos:

$f(x) = -x^3 + 3x^2 -2x \Rightarrow f(1) = 0$

$f'(x) = -3x^2 + 6x - 2 \Rightarrow f'(1) = 1$

$f''(x) = -6x + 6 \Rightarrow f''(1) = 0$

```python
# Dois mil pontos entre 0 e 2:
interval = np.arange(0, 2, 0.001)

def f(x):
    return -np.power(x,3) + 3*np.power(x,2) - 2*x

def df(x):
    return -3*np.power(x,2) + 6*x -2

def ddf(x):
    return -6*x + 6

fig, ax = plt.subplots()

# Plots:
plt.plot(interval, f(interval), label="f(x) Polinomial", linewidth=3)
plt.plot(interval, df(interval), label="f'(x) Polinomial", linewidth=3)
plt.plot(interval, ddf(interval), label="f''(x) Polinomial", linewidth=3)

# Grid:
ax.set_axisbelow(True)
ax.minorticks_on()
ax.grid(which='major', linestyle='-', linewidth='0.5', color='red')
ax.grid(which='minor', linestyle=':', linewidth='0.5', color='black')

# Legend:
plt.legend()
```

### Terceiro Caso (polinomial)

Para esse caso, temos:

$f(x) = - \frac{1}{2} \left( x -1 \right) \left( x - 3\right) \Rightarrow f(1) = 0$

$f'(x) = - \frac{1}{2} \left( 2x - 4 \right) = -x + 2 \Rightarrow f'(1) = 1$

$f''(x) = -1 \Rightarrow f''(x) = -1 < 0$

```python
# Dois mil pontos entre 0 e 2:
interval = np.arange(0, 2, 0.001)

def f(x):
    return -(1/2)*(x-1)*(x-3)

def df(x):
    return -(1/2)*(2*x-4)

def ddf(x):
    return  np.array([-1 for i in x])

fig, ax = plt.subplots()

# Plots:
plt.plot(interval, f(interval), label="f(x) Polinomial", linewidth=3)
plt.plot(interval, df(interval), label="f'(x) Polinomial", linewidth=3)
plt.plot(interval, ddf(interval), label="f''(x) Polinomial", linewidth=3)

# Grid:
ax.set_axisbelow(True)
ax.minorticks_on()
ax.grid(which='major', linestyle='-', linewidth='0.5', color='red')
ax.grid(which='minor', linestyle=':', linewidth='0.5', color='black')

# Legend:
plt.legend()
```

### Primeiro Caso (transcendental)

Para esse caso temos:

$ f(x) = e^{x-1} -1 \Rightarrow f(1) = 0 $

$ f'(x) = e^{x-1} \Rightarrow f'(1) = 1$

$ f''(x) = e^{x-1} \Rightarrow f''(1) = 1 > 0$

```python
# Dois mil pontos entre 0 e 2:
interval = np.arange(0, 2, 0.001)

def f(x):
    return np.exp(x-1)-1

def df(x):
    return np.exp(x-1)

def ddf(x):
    return  np.exp(x-1)

fig, ax = plt.subplots()

# Plots:
plt.plot(interval, f(interval), label="f(x) Polinomial", linewidth=3)
plt.plot(interval, df(interval), label="f'(x) Polinomial", linewidth=3)
plt.plot(interval, ddf(interval), label="f''(x) Polinomial", linewidth=3)

# Grid:
ax.set_axisbelow(True)
ax.minorticks_on()
ax.grid(which='major', linestyle='-', linewidth='0.5', color='red')
ax.grid(which='minor', linestyle=':', linewidth='0.5', color='black')

# Legend:
plt.legend()
```

### Segundo Caso (transcendental)

Para esse caso temos:

$f(x) = sen(x-1) \Rightarrow f(1) = 0$

$f'(x) = cos(x-1) \Rightarrow f'(1) = 1$

$f''(x) = -sen(x-1) \Rightarrow f''(1) = 0$

```python
# Dois mil pontos entre 0 e 2:
interval = np.arange(0, 2, 0.001)

def f(x):
    return np.sin(x-1)

def df(x):
    return np.cos(x-1)

def ddf(x):
    return  -np.sin(x-1)

fig, ax = plt.subplots()

# Plots:
plt.plot(interval, f(interval), label="f(x) Polinomial", linewidth=3)
plt.plot(interval, df(interval), label="f'(x) Polinomial", linewidth=3)
plt.plot(interval, ddf(interval), label="f''(x) Polinomial", linewidth=3)

# Grid:
ax.set_axisbelow(True)
ax.minorticks_on()
ax.grid(which='major', linestyle='-', linewidth='0.5', color='red')
ax.grid(which='minor', linestyle=':', linewidth='0.5', color='black')

# Legend:
plt.legend()
```

### Terceiro Caso (transcendental)

Para esse caso temos:

$ f(x) = ln(x) \Rightarrow f(1) = 0 $

$ f'(x) = \frac{1}{x} \Rightarrow f'(1) = 1 $

$ f''(x) = - \frac{1}{x^2} \Rightarrow f''(1) = -1 < 0 $


```python
# Mil pontos entre 0.5 e 1,5:
interval = np.arange(0.5, 1.5, 0.001)

def f(x):
    return np.log(x)

def df(x):
    return 1/x

def ddf(x):
    return  -1/np.power(x,2)

fig, ax = plt.subplots()

# Plots:
plt.plot(interval, f(interval), label="f(x) Polinomial", linewidth=3)
plt.plot(interval, df(interval), label="f'(x) Polinomial", linewidth=3)
plt.plot(interval, ddf(interval), label="f''(x) Polinomial", linewidth=3)

# Grid:
ax.set_axisbelow(True)
ax.minorticks_on()
ax.grid(which='major', linestyle='-', linewidth='0.5', color='red')
ax.grid(which='minor', linestyle=':', linewidth='0.5', color='black')

# Legend:
plt.legend()
```
