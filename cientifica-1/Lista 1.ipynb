## Questão 1 .1 - Resposta:

Podemos considerar uma função $f$, e uma tangente a ela, conforme a figura abaixo:


Pela figura, vemos que a tangente passa por $f$ no ponto $(x_0, f(x_0) = y_0)$ e que o ponto $x_1$ pode ser encontrado por:

$$
\frac{\Delta y}{\Delta x} = \frac{y_0 - 0}{x_0 - x_1} = f'(x_0) \Rightarrow x_1 = x_0 - \frac{f(x_0)}{f'(x_0)}
$$

Como temos que aplicar reiteiradamente o método, até que a diferença de $x_i$ para $x_{i+1}$ e de $f(x_i)$ e $0$ sejam menores que os seus respectivos erros, temos que:

$x_{i+1} = x_i - \frac{f(x_i)}{f'(x_i)}$, até que $\lvert x_{i+1} - x_{i} \rvert < \varepsilon_y$ ou $\lvert f(x_i) \rvert < \varepsilon_y$, para qualquer $f$.

Mas vamos assumir que $g(x) = - f(x) \Rightarrow g'(x) = -f'(x)$ e, como a formula de Aproximação de Newton pode ser usada para qualquer $f$, temos que, para $g$:

$x_{i+1} = x_i - \frac{g(x_i)}{g'(x_i)} = x_i - \frac{-f(x_i)}{-f'(x_i)} = x_i - \frac{(-1) \cdot f(x_i)}{(-1) \cdot f'(x_i)} =  x_i - \frac{f(x_i)}{f'(x_i)}$

Que é exatamente as fórmulas de iteração de Newton para $f$, provando o argumento.
